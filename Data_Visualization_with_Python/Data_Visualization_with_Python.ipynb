```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_json("rain.json")
print(df.head())
```

```python
print("Dataframe statistics: ", df.describe)
```

```python
df.plot(x='Month', y='Temperature')
plt.grid(True)
plt.legend()

df.plot(x='Month', y='Rainfall')
# Adding grid lines
plt.grid(True)

# Adding a legend
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot( df['Month'], df['Temperature'], label='Temperature')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot( df['Month'], df['Rainfall'], label='Rainfall')

plt.title('Monthly Rainfall')
plt.xlabel('Month')
plt.ylabel('Rainfall (mm)')

plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot( df['Month'], df['Rainfall'], label='Rainfall')
plt.plot( df['Month'], df['Temperature'], label='Temperature')

plt.grid(True)
plt.legend()

plt.show()
```

```python
df2 = pd.read_csv(r"tempYearly.csv")
df2.head()
```

```python
sns.set(rc={'figure.figsize':(12,6)})
sns.scatterplot(x='Year', y='Temperature', data=df2)
plt.show()
```

```python
sns.regplot(x='Year', y='Temperature', data=df2)
plt.show()
```

```python
data4 = pd.read_csv(r"tempYearly.csv")
data4.head()
```

```python
sns.jointplot(x="Rainfall", y="Temperature", data=data4)
plt.show()

# scatterplot in the middle and histograms on the side
# useful when trying to see quantities on each axis
```

```python
sns.jointplot(x="Rainfall", y="Temperature", data=data4, kind="hex")
plt.show()
```

```python
sns.jointplot(x="Rainfall", y="Temperature", data=data4, kind="reg")
plt.show()
```

Linear correlation of 2 datasets using Seaborn: regplot

Advantage of jointplot: allows you to see 2 plot types at the same time
