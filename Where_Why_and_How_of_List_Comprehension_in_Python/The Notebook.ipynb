### Task 2 
    There are three main applications of list comprehension:
        1 - List Manipulation 
        2 - List Filtering
        3 - Extracting Information

### 1- List Manipulation
        
        When we have an input list and we want to generate an output list from the inputlist. In other words,
        changing the values input list and store them in the output list. 

Let's take a look at this example.
    

Example 1: the goal is to multiply each element in input list by 2 and generate an output list containing the multiplied numbers

```python
input_list = [12,10,9,18,20,17]
output_list = [] 
for number in input_list:
    output_list.append(number * 2)

output_list
```

### Now Let's use list comprehension instead
<img src = "images/list_comprehension.png">

```python
input_list = [12,10,9,18,20,17]
output_list = [number * 2 for number in input_list]
output_list
```

Let's do another example: devide every value by ten and put in the output list 

```python
input_list = [20,100,33,89,99,10]
output_list = [number / 10 for number in input_list]
output_list
```

Exercise: We have list of names. use list comprehension to generate a list of the length of each name (number of characters in each name).

### Task 3 
    In this task we are going to talk about the second application of the list comprehension
### 2 - List Filtering
    We can combine if else statement with list comprehension



<img src = "images/list_comprehension_if_else.png">

Let's filter out the values greater than 50

```python
input_list = [20,100,33,89,99,10]
numbers_greater_than_50 = [number for number in input_list if number > 50]
numbers_greater_than_50
```

Let's do another example: we have a list of names, filter out only those names starting with an 'A' or 'a'

```python
names = ["Ipek","Alexia","Maria","Chris","alice","Minh","Joey","Archy",'amir','Kim','Aline']
names_with_a_or_A = [name for name in names if name[0] == "A" or name[0] == "a"]
names_with_a_or_A
```

As you can see we could combine different conditions in the if statement

Exercise - We have list of names, filter out only those names which are longer than 4 letters

```python
names = ["Ipek","Alexia","Maria","Chris","alice","Minh","Joey","Archy",'amir','Kim','Aline']
names_with_4_or_more_chars = [name for name in names if len(name) > 4]
names_with_4_or_more_chars
```

We have a list of numbers, use list comprehension to multiply the odd number by 2 and even values by 3

```python
numbers = [12,10,9,18,20,17]
modify_numbers = [number*3 if number % 2 == 0 else number * 2 for number in numbers ]
modify_numbers
```

### Extracting Information 
    We can Also apply list comprehension to a list of dictionary objects to extract information. 

Assume that we have a list of users. for each user we have stored the name and the age inside a dictionary. 

Let's use list comprehension to extract the list of the names. 

```python
users = [ 
    
    {"Name": "Alexia" , "Gender" : "Female" , "Age" : 20},
    {"Name": "Cina" ,"Gender" : "Male", "Age" : 39},    
    {"Name": "John" , "Gender" : "Male" , "Age" : 23},
    {"Name": "Martina" ,"Gender" : "Female", "Age" : 45},    
    {"Name": "Marvin" ,"Gender" : "Male", "Age" : 52},
    {"Name": "Kim" ,"Gender" : "Female", "Age" : 29},
    {"Name": "Aline" ,"Gender" : "Female", "Age" : 31},
    {"Name": "Layla" ,"Gender" : "Female", "Age" : 50},
    {"Name": "Sara" ,"Gender" : "Female", "Age" : 24},

]

objects_in_users = [object for object in users]
objects_in_users
```

```python
names_objects = [object["Name"] for object in users]
names_objects
```

Now what if we want a list of users whose age is higher than 30

```python
names_objects = [object["Name"] for object in users if object["Age"] > 30 ]
names_objects
```

Exercise - Filter out all the users who are female

```python
names_female = [object["Name"] for object in users if object["Gender"] != "Female" ]
names_female
```

```python
names_male = [object["Name"] for object in users if object["Gender"] != "Male" ]
names_male
```

We have a nested list called user groups. Let's extract the users older than 30 using List Comprehension

```python
user_groups = [ 
    
    [
        {"Name": "Alexia" , "Gender" : "Female" , "Age" : 20},
        {"Name": "Cina" ,"Gender" : "Male", "Age" : 39},    
        {"Name": "John" , "Gender" : "Male" , "Age" : 23},
        {"Name": "Martina" ,"Gender" : "Female", "Age" : 45}
    ],
    
    [ 
        {"Name": "Marvin" ,"Gender" : "Male", "Age" : 52},
        {"Name": "Marta" ,"Gender" : "Female", "Age" : 29},
        {"Name": "Aline" ,"Gender" : "Female", "Age" : 31},
        {"Name": "Mina" ,"Gender" : "Female", "Age" : 50}
    ]


]
```

```python
users_over_30 = [user for group in user_groups for user in group if user["Age"] > 30 ]
users_over_30 
```

So we can use nested loops in list comprehension.

Exercise: Use nested loops and list comprehension to extract all the user's whose name begins with an 'M' or 'm'

```python
users_with_m_or_M = [user for group in user_groups for user in group if user["Name"][0] == "M" or user["Name"][0] == "m"]
users_with_m_or_M 
```

### Task 5
We have a dataset containing the data about different UFO (Unidentified flying object) observations, reported by the civilians in different cities around the globe. 


Importing the modules

```python
import json
import matplotlib.pyplot as plt
from collections import Counter
from statistics import mean
```

Loading the data

```python
with open('dataset.json') as file:
    data = json.load(file)
data
```

```python
data[:2]
```

```python
data[0]["city"]
```

How many observations do we have?

```python
len(data)
```

What are the most common UFO shapes?

```python
shapes = [object["shape"] for object in data ]
shapes
```

```python
most_common_shape = dict(Counter(shapes))
most_common_shape
```

Let's visualize the results using a bar chart

```python
x = most_common_shape.keys()
x
```

```python
y = most_common_shape.values()
y
```

```python
plt.rcParams.update({"font.size": 18})
plt.figure(figsize= (25,10))
plt.xticks(rotation= 90)
plt.bar(x = x, height = y)
```

Let's calculate the average duration of observations (in minutes) with the shape of disk

```python
data[0]
```

```python
durations = [float(obj["duration (seconds)"])/60 for obj in data if obj["shape"] == "disk"]
durations
```

```python
mean(durations) 
```

Find the year with most observations

```python
data[0]
```

```python
def extract_year(obj):
    date = obj["datetime"].split(" ")[0]
    if len(date.split("/")) == 3:
        year = date.split("/")[-1]
        return year
    else:
        return ""

years = [extract_year(obj) for obj in data]
years
```

```python
years_count = dict(Counter(years))
years_count
```

```python
# removing invalid years:
del years_count['']
years_count
```

Let's sort the dictionary by keys

```python
years_sorted = dict(sorted(years_count.items()))
years_sorted
```

Let's visualize the results

```python
x = list(years_sorted.keys())
y = list(years_sorted.values())

plt.rcParams.update({"font.size": 18})
plt.figure(figsize= (25,10))
plt.xticks(rotation= 45)
plt.plot(x,y)
```
