## T tests

```python
import scipy.stats as stats
```

```python
a = st_data_2009_clean['Patent applications, residents']
b = st_data_2018_clean['Patent applications, residents']

# independent t test
stats.ttest_ind(a,b)
```

```python
for i in range(1,len(columns)):
    a = st_data_2009_clean[columns[i]]
    b = st_data_2018_clean[columns[i]]
    print(columns[i])
    statistic, pvalue = stats.ttest_ind(a,b)
    print("Statistic: %s p-value: %s" %(statistic, pvalue))
    print("\n")
```

```python
for i in range(1,len(columns)):
    a = st_data_2009_clean[columns[i]]
    b = st_data_2018_clean[columns[i]]
    print(columns[i])
    statistic, pvalue = stats.ttest_ind(a,b)
    print("Statistic: %s p-value: %s" %(statistic, pvalue))
    if pvalue < 0.05:
        print("Significant")
    else:
        print("Not Significant")
    print("\n")
```

Because our 2 datasets had 2 different number of countries in them we actually ran independent T tests.

It would be useful to run a relative t test comparing the 2 years together where we would simply switch from saying stats.ttest_ind to stats.ttest_rel but for that to work you need to be comparing exactly the same countries in exactly the same years over the same statistic.
